In this notebook, we will look into how we can process the audio to detect the pitch. We will look into all the algorithms we can use analyse the best algorithm for our case. We will also visualize the algorithms as much as possible.

```python
#Imports
import numpy as np
import IPython
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.fft import fft
from scipy import signal
```

```python
#Reading Sound File
audioPath = './data/flutter_sound_example.wav'
samplerate, data = wavfile.read(audioPath)
data = data[0: samplerate]
IPython.display.Audio(audioPath)
# print(samplerate)
```

```python
# print(data)
plt.plot(data)
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.show()
```

```python
window = signal.windows.hann(len(data))
plt.plot(window)
plt.title("Hann window")
plt.ylabel("Amplitude")
plt.xlabel("Sample")
plt.show()
```

```python
windowed_data = [data[i] * window[i] for i in range(len(data))]
plt.plot(windowed_data)
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.show()
```

```python
fft_data = fft(windowed_data)
plt.plot(np.abs(fft_data))
plt.show()
```

```python
#  Performing Yin
buffer_size = 2048
threshold = 0.001
audio_signal = windowed_data - np.mean(windowed_data)
# plt.plot(np.abs(audio_signal))
# plt.show()
audio_signal = np.square(audio_signal)
yin_buffer = np.zeros(buffer_size//2)
for tau in range(1, buffer_size//2):
        for i in range(buffer_size//2):
            delta = audio_signal[i] - audio_signal[i + tau]
            yin_buffer[tau] += delta * delta
        yin_buffer[tau] = yin_buffer[tau] / (buffer_size//2)
        
plt.plot(yin_buffer);
plt.show()
        
# Perform cumulative mean normalized difference
running_sum = 0
yin_result = np.zeros(buffer_size//2)
for tau in range(1, buffer_size//2):
    running_sum += yin_buffer[tau]
    yin_result[tau] = running_sum / tau / 0.000001
    
plt.plot(yin_result);
plt.show()

pitch_index = np.argmin(yin_result[1:]) + 1 # Find the minimum YIN result and add 1
pitch = samplerate / pitch_index # Calculate the pitch in Hz
if yin_result[pitch_index] < threshold: # Check if the minimum YIN result is below the threshold
    print(pitch)
else:
    print(-1) # Pitch could not be estimated
```
